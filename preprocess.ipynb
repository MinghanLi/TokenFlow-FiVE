```python
import os
import torch
import matplotlib.pyplot as plt
```

```python
dir_path = '/PHShome/ml1833/code/TokenFlow/latents/sd_2.1/wolf/steps_500/nframes_40/latents'
filename = 'noisy_latents_1.pt'

file_path = os.path.join(dir_path, filename)
data = torch.load(file_path)
print(data.shape)

for t in range(0, data.shape[0], 8):
    sample = data[t].cpu()  # Shape [4, 64, 64]

    # Plot each of the 4 channels separately
    fig, axes = plt.subplots(1, 4, figsize=(12, 6))
    for i in range(4):
        axes[i].imshow(sample[i], cmap='gray')
        axes[i].set_title(f'Channel {i+1}')
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()
```
